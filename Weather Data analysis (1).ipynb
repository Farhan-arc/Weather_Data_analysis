```python
# How to Plot a Histogram of 'Wind Speed' Using Matplotlib ?
```

```python
plt.figure(figsize=(10, 6))
plt.hist(Weather['Wind Speed_km/h'], bins=30, edgecolor='black', color='skyblue')
plt.title('Distribution of Wind Speed', fontsize=12)
plt.xlabel('Wind Speed (km/h)', fontsize=12)
plt.ylabel('Frequency', fontsize=12)
plt.grid(True)
plt.show()
```
